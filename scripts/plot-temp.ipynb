```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
data = np.loadtxt('./output/old-and-processed/t0', comments='#', delimiter=',')

# Create the figure and primary axis
fig, ax1 = plt.subplots()

# Plot the bottom x-axis and left y-axis (Distance vs Depth)
scatter1 = ax1.scatter(x=data[:, 1], y=data[:, 2], c=data[:, 4], label='Distance vs Depth', cmap='viridis')

# Customize the primary axes
ax1.set_xlabel('Distance (Km)')
ax1.set_ylabel('Depth (Km)')
ax1.invert_yaxis()  # Invert y-axis for depth
ax1.tick_params(axis='y')

# Add a colorbar for the first scatter
cbar1 = plt.colorbar(scatter1, ax=ax1, pad=0.15)
cbar1.set_label('Temperature (C)')

# Create a secondary y-axis for Pressure
ax2 = ax1.twinx()
ax2.set_ylabel('Pressure (GPa)')  # Update label to GPa
ax2.set_ylim(data[:, 3].min() * 1e-4, data[:, 3].max() * 1e-4)  # Set y-limits based on converted Pressure data
ax2.invert_yaxis()  # Invert the pressure axis

# Create a secondary x-axis at the top for Vs
ax3 = ax1.twiny()
ax3.set_xlabel('Vs (km/s)')

# Set the limits of ax3 to correspond with Vs data
ax3.set_xlim(data[:, 7].min(), data[:, 7].max())  # Set limits based on Vs data

# Adjust the layout to prevent overlap
fig.tight_layout()

# Show the plot
plt.show()
```
