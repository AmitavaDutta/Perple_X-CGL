```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load the data
filename = 'HP02-dry-Arch-Basalt.tab'
data = np.loadtxt(filename, comments='#')

# Plotting
plt.scatter(x=data[:, 0] - 273.15, y=data[:, 1] / 1000, c=data[:, 4], cmap='inferno')  # Temperature in °C, Pressure in GPa
plt.ylim(100, 0)  # Inverted y-axis for pressure
plt.colorbar(label='Vs (km/s)')  # Color bar for Vs values
plt.xlabel('Temperature (°C)')  # X-axis label
plt.ylabel('Pressure (GPa)')  # Y-axis label
plt.title(os.path.splitext(os.path.basename(filename))[0])  # Title without .tab extension
plt.show()
```
